```python
import random
```

```python

count = 0
for x in range(1,11):
    a = random.randint(1,50)
    b = random.randint(1,50)
    print('%d + %d = ' %(a,b))
    c = input()
    if c.isdecimal():
        c=int(c)
        if c==a+b:
            print('정답!')
            count +=1
        else:
            print('오답!')
    else:
        print('숫자로 입력하세요')
print('{}개 맞음'.format(count))
```

```python
count = 0
oper = ['+','-','*','/']
for i in range(1,11):
    a = random.randint(1,50)
    b = random.randint(1,50)
    op = random.choice(oper)
    quiz =str(a)+op+str(b)
    print(quiz,'=')
    result = input('답은?')
    result = int(result)
    if result == eval(quiz):
        print('정답!')
        count += 1
    else:
        print('오답!')
print('{}개 맞음'.format(count))
```
